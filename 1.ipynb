```python
import pandas as pd

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def primes_up_to_300():
    primes = []
    for num in range(1, 301):
        if is_prime(num):
            primes.append(num)
    return primes

prime_numbers = primes_up_to_300()
df = pd.DataFrame(prime_numbers, columns=["Prime Numbers"])

print(df)
```

```python
import pandas as pd

def generate_fibonacci(max_value):
    fib_series = [0, 1]

    while True:
        next_value = fib_series[-1] + fib_series[-2]
        if next_value > max_value:
            break
        fib_series.append(next_value)

    fib_series = [num for num in fib_series if num >= 1]

    return fib_series
max_value = 100

fib_series = generate_fibonacci(max_value)
df = pd.DataFrame(fib_series, columns=["Fibonacci Numbers"])

print(df)
```

```python
def calculate_sum(numbers):
    total = 0
    for number in numbers:
        total += number
    return total

def find_maximum(numbers):
    max_num = numbers[0]
    for number in numbers:
        if number > max_num:
            max_num = number
    return max_num

def find_minimum(numbers):
    min_num = numbers[0]
    for number in numbers:
        if number < min_num:
            min_num = number
    return min_num
user_input = input("Enter numbers separated by spaces: ")
numbers = [int(num) for num in user_input.split()]

sum_of_numbers = calculate_sum(numbers)
maximum_number = find_maximum(numbers)
minimum_number = find_minimum(numbers)
print("SUM OF NUMBERS: ")
print(sum_of_numbers)
print("MAXIMUM NUMBER: ")
print(maximum_number)
print("MINIMUM NUMBER: ")
print(minimum_number)
```
