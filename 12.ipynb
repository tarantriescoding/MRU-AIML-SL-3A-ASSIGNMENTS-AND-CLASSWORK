```python
#The predefined columns are -

#Passenger ID - To identify unique passengers
#Survived - If they survived or not
#P Class - The class passengers travelled in
#Name - Passenger Name
#Sex - Gender of Passenger
#Age - Age of passenger
#SibSp - Number of siblings or spouse
#Parch - Parent or child
#Ticket - Ticket number
#Fare - Amount paid for the ticket
#Cabin - Cabin of residence
#Embarked - Point of embarkment
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.head()
```

```python
# Dropping those rows where taget variable is missing

df = df.dropna(subset = ['Survived'])
```

```python
df.shape
```

```python
df.info()
```

```python
x = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df["Survived"]
```

```python
x.info()
```

```python
# Label encoding
le = LabelEncoder()
x['Sex']= le.fit_transform(x['Sex'])
```

```python
x.head()
```

```python
x['Age'] = x['Age'].fillna(x['Age'].mean())
```

```python
x['Age']
```

```python
x.info()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                test_size = 0.2,
                                                  random_state=42)
```

```python
# Create a Random Forest Classifier
# n_estimators = 100 decision Trees
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the classifier
model.fit(x_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```
