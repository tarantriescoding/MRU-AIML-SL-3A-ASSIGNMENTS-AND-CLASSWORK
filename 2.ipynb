```python
import matplotlib.pyplot as plt
import numpy as np
def plotfunction(func):
  x1= np.linspace(-10,10,1000)
  # y = x**2
  y1 = [func(x) for x in x1]
  print("x values are : ", x1)
  print("y values are : ", y1)
  plt.plot(x1,y1)
  plt.show()

plotfunction(lambda x: x**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt


coefficients = list(map(float, input("Enter the coefficients a, b, and c separated by space: ").split()))
a, b, c = coefficients

x = np.linspace(-10, 10, 400)

y = a * x**2 + b * x + c

h = 1e-7
y_derivative = (a * (x + h)**2 + b * (x + h) + c - (a * x**2 + b * x + c)) / h

fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(x, y)
ax[0].set_title('Cost Function')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].plot(x, y_derivative)
ax[1].set_title("Derivative of Cost Function")
ax[1].set_xlabel('x')
ax[1].set_ylabel("y'")

plt.tight_layout()
plt.show()
```

```python
def plot_function(func):
  x_values = np.linspace(-5, 5, 1000)
  delta_x = 0.0001
  y_values = (func(x_values +delta_x) - func(x_values)) / delta_x
  plt.plot(x_values, y_values)
  plt.show()

plot_function(lambda x: x**2)
```

```python
import numpy as np
def plot_function(func):
  x_values = np.linspace(-5, 5, 1000)
  delta_x = 0.0001
  y_values = np.gradient(func(x_values), delta_x)
  plt.plot(x_values, y_values)
  plt.show()

plot_function(lambda x: x**2)
```

```python
def gradient_decent(func,w):
  list_of_weights = []
  weight = w
  delta = 0.0001
  for i in range (100):
    derivative = (func(weight+delta) - func(weight))/delta
    weight = weight - derivative
    list_of_weights.append(weight)
    return list_of_weights
gradient_descent(lambda x: x**2+4*x+3, 10)
```
