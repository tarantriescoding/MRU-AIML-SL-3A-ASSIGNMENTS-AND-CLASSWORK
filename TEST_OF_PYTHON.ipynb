```python
# write a program to find the minimum using gradient descent and plot the graph also
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def gradient(x):
    return 2*x

learning_rate = 0.1
initial_x = 10
num_iterations = 100
x = initial_x


for i in range(num_iterations):
    grad = gradient(x)
    x = x - learning_rate * grad
    print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")

print(f"Minimum value of f(x) is approximately at x = {x}, f(x) = {f(x)}")
x_range = np.linspace(-11, 11, 400)
plt.plot(x_range, f(x_range), label='f(x) = x^2', color='blue')
plt.title('Gradient Descent Optimization')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

```python
#wap to predict the price of a bike using linear regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


bike_data = pd.read_csv('Bike_Price.csv')
bike_data.head()
```

```python
# PLEASE WORK
bike_data_clean = bike_data.drop(columns=['S.no', 'Bike_model'])
bike_data_clean = pd.get_dummies(bike_data_clean, columns=['Bike_company', 'Engine_type', 'Fuel_type'])
bike_data_clean['CC(Cubic capacity)'] = bike_data_clean['CC(Cubic capacity)'].str.replace('CC', '').astype(float)
bike_data_clean['Fuel_Capacity'] = pd.to_numeric(bike_data_clean['Fuel_Capacity'].str.extract('(\d+)')[0], errors='coerce')
bike_data_clean.dropna(inplace=True)
X = bike_data_clean.drop(columns=['Price'])
y = bike_data_clean['Price']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=169)
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
r_squared = model.score(X_test, y_test)
mse = mean_squared_error(y_test, y_pred)
print("R-SQUARED SCORE: " + str(r_squared))
print("MEAN_SQUARED ERROR: " + str(mse))
```
